```python
def a(n):
  if n==0 or n==1:
    return 1
  else:
    return n + a(n-1)

print(a(3))
print(a(5))
```

當設定n=5時，因為5不是0與1，所以會在n+a(n-1)跑。
因為有定義 a(n)，所以在確認a(n-1)時會重新run過。
故會一直重新跑，直到a(n-1)的數字是0跑到if函數才停止。
因此n=5的答案出來會是5+4+3+2+1+0=15
